```python
class NodoBinario:
    def __init__( self, value, left = None, right = None ):
        self.data = value
        self.left = left 
        self.right = right 

class BinarySearchTree:
    def __init__( self ):
        self.root = None

    def insert( self, value ):
        if self.root == None: 
            self.root = NodoBinario( value )

        else:
            self.__insert_nodo( self.root, value )

    def __insert_nodo( self, nodo, value):
        if nodo.data == value:
            pass
        elif value < nodo.data: #Si se cumple, ir a la izquierda
            if nodo.left == None:
                nodo.left = NodoBinario( value ) 
            else:
                self.__insert_nodo( nodo.left, value )

        else: #Si no se cumplió, ir a la derecha
            if nodo.right == None:
                nodo.right = NodoBinario( value ) 
            else:
                self.__insert_nodo( nodo.right, value )

    def transversal( self, formato = "posorden"):
        if formato == "posorden":
            self.recorrido_pos( self.root )

        elif formato == "preorden":
            self.recorrido_pre( self.root )
        else: #inorden
            self.recorrido_in( self.root )
        print()

    def recorrido_pos( self, nodo ):
        if nodo != None:
            self.recorrido_pos( nodo.left )
            self.recorrido_pos( nodo.right )
            print( str(nodo.data), end= " , " )

    def recorrido_pre( self, nodo ):
        if nodo != None:
            print( str(nodo.data), end= " , " )
            self.recorrido_pre( nodo.left )
            self.recorrido_pre( nodo.right )
    
    def recorrido_in( self, nodo ):
        if nodo != None:
            self.recorrido_in( nodo.left )
            print( str(nodo.data), end= " , " )
            self.recorrido_in( nodo.right )

    def search( self, value ):
        if self.root == None:
            print("El árbol está vacío")
            return None 
        else:
            self.__search( self.root, value )
    
    def __search( self, nodo, value ):
        if nodo == None:
            print("El dato no existe en el árbol")
            return None
        elif nodo.data == value:
            print(f"Encontrado --> {nodo.data}")
            return nodo 
        elif value < nodo.data:
            return self.__search( nodo.left, value )
        else:
            return self.__search( nodo.right, value )
```

```python
class BinaryNode:
    def __init__( self , value , left = None , right = None, father = None ):
        self.data = value
        self.left = left
        self.right = right
        self.father = father
 
class BinarySearchTree:
    def __init__( self ):
        self.root = None
    
    def insert( self , value ):
        if self.root == None:
            self.root = BinaryNode( value )
        else:
            self.__insert_nodo( self.root , value )
        
    def __insert_nodo(self , nodo , value ):
        if nodo.data == value: 
            pass
        elif value < nodo.data:# ir a la izq
            if nodo.left == None:
                nodo.left = BinaryNode( value, father = nodo )
            else:
                self.__insert_nodo( nodo.left , value )
        else:  # ir a la derecha
            if nodo.right == None:
                nodo.right = BinaryNode( value, father = nodo )
            else:
                self.__insert_nodo( nodo.right , value )
    
    def transversal( self , formato = "posorden"):
        if formato == "posorden":
            self.recorrido_pos( self.root )
        elif formato == "preorden":
            self.recorrido_pre(self.root )
        else: # inorden
            self.recorrido_in( self.root )
        print()
 
    def  recorrido_pos( self , nodo ):
        if nodo != None:
            self.recorrido_pos( nodo.left )
            self.recorrido_pos( nodo.right )
            print( str(nodo.data) , end=" , " )
    
    def  recorrido_pre( self , nodo ):
        if nodo != None:
            print( str(nodo.data) , end=" , " )
            self.recorrido_pre( nodo.left )
            self.recorrido_pre( nodo.right )
    
    def  recorrido_in( self , nodo ):
        if nodo != None:
            self.recorrido_in( nodo.left )
            print( str(nodo.data) , end=" , " )
            self.recorrido_in( nodo.right ) 
 
    def search( self , value ):
        if self.root == None:
            print("El árbol está vacío")
            return None
        else:
            return self.__search( self.root , value )
 
    def __search( self , nodo , value ):
        if nodo == None:
            print("El dato no existe en el árbol") 
            return None
        elif nodo.data == value:
            print(f"Encontrado: | {nodo.data} |")
            return nodo
        elif value < nodo.data:
            return self.__search( nodo.left , value  )
        else:
            return self.__search( nodo.right , value  )  

    def reemplazo( self, nodo ):
        if nodo.left.right:
            return "mayor_del_menor"

        elif nodo.right.left:
            return "menor_del_mayor"
        
        else:
            return "ninguno"

    def remove( self, value ):
        nodo = self.search( value )
        nodo_padre = nodo.father
        #Caso 1: El nodo a borrar no tiene hijos (es una hoja).
        if nodo.left == None and nodo.right == None:
            print(f"| {nodo.data} | Eliminado")
            if nodo.data < nodo.father.data:
                nodo.father.left = None
            else:
                nodo.father.right = None

        #Caso 2: Tiene un solo hijo
        elif (nodo.left and nodo.right == None) or (nodo.left == None and nodo.right) : 
            if nodo.data < nodo.father.data: #Si se cumple entonces "nodo" es hijo izquierdo de father
                if nodo.left: #Si del nodo que queremos borrar existe el hijo izquierdo
                    nodo.father.left = nodo.left
                    print(f"| {nodo.data} | Eliminado")
                else: #Si no, existe el hijo derecho
                    nodo.father.left = nodo.right
                    print(f"| {nodo.data} | Eliminado")
            else: #Si no, entonces es hijo derecho
                if nodo.left:
                    nodo.father.right = nodo.left
                    print(f"| {nodo.data} | Eliminado")
                else:
                    nodo.father.right = nodo.right
                    print(f"| {nodo.data} | Eliminado")
        
        #Caso 3: Tiene ambos hijos (y son sub-árboles)
        else:
            caso = self.reemplazo( nodo )
            copia_left = nodo.left
            copia_right = nodo.right
            #Caso 3.1: El nodo a borrar tiene un mayor de los menores 
            if caso == "mayor_del_menor":
                #Caso 3.1.1 El nodo a borrar es la raiz
                if nodo == self.root:
                    new_root = self.root.left.right
                    self.root = new_root
                    self.root.left = None
                    self.root.right = None
                    self.root.left = copia_left 
                    self.root.right = copia_right
                    self.root.left.right = None
                    print(f"| {nodo.data} | Eliminado")
                
                #Caso 3.1.2 El nodo a borrar no es la raíz
                else:
                    new_nodo = nodo.left.right
                    if nodo.data < nodo.father.data:
                        new_nodo = nodo.father.left.left.right
                        nodo.father.left = new_nodo
                        nodo.father.left.left = None
                        nodo.father.left.right = None
                        nodo.father.left.left = copia_left
                        nodo.father.left.right = copia_right
                        nodo.father.left.left.right = None 
                        print(f"| {nodo.data} | Eliminado")
                    else:
                        new_nodo = nodo.father.right.left.right
                        nodo.father.right = new_nodo
                        nodo.father.right.left = None
                        nodo.father.right.rigth = None
                        nodo.father.right.left = copia_left
                        nodo.father.right.rigth = copia_right
                        nodo.father.right.left.right = None 
                        print(f"| {nodo.data} | Eliminado")

            #Caso 3.2: El nodo a borrar tiene un menor de los mayores
            elif caso == "menor_del_mayor":
                #Caso 3.2.1 El nodo a borrar es la raiz
                if nodo == self.root:
                    new_root = self.root.right.left
                    self.root = new_root
                    self.root.left = None
                    self.root.right = None
                    self.root.left = copia_left 
                    self.root.right = copia_right
                    self.root.right.left = None
                    print(f"| {nodo.data} | Eliminado")
                
                #Caso 3.1.2 El nodo a borrar no es la raíz
                else:
                    new_nodo = nodo.right.left
                    if nodo.data < nodo.father.data:
                        new_nodo = nodo.father.left.right.left
                        nodo.father.left = new_nodo
                        nodo.father.left.left = None
                        nodo.father.left.right = None
                        nodo.father.left.left = copia_left
                        nodo.father.left.right = copia_right
                        nodo.father.left.right.left = None 
                        print(f"| {nodo.data} | Eliminado")
                    else:
                        new_nodo = nodo.father.right.right.left
                        nodo.father.right = new_nodo
                        nodo.father.right.left = None
                        nodo.father.right.right = None
                        nodo.father.right.left = copia_left
                        nodo.father.right.right = copia_right
                        nodo.father.right.right.left = None 
                        print(f"| {nodo.data} | Eliminado")

            #Caso 3.3: El nodo a borrar no tiene ni un mayor de los menores ni un menor de los mayores
            else:
                #Caso 3.3.1 El nodo a borrar es la raiz
                if nodo == self.root:
                    new_root = self.root.right
                    self.root = new_root
                    self.root.left = None
                    self.root.right = None
                    self.root.left = copia_left 
                    self.root.right = copia_right
                    self.root.right = None
                    print(f"| {nodo.data} | Eliminado")
                
                #Caso 3.3.2 El nodo a borrar no es la raíz
                else:
                    new_nodo = nodo.right
                    if nodo.data < nodo.father.data:
                        new_nodo = nodo.father.left.right
                        nodo.father.left = new_nodo
                        nodo.father.left.left = None
                        nodo.father.left.right = None
                        nodo.father.left.left = copia_left
                        print(f"| {nodo.data} | Eliminado")
                    else:
                        new_nodo = nodo.father.right.right
                        nodo.father.right = new_nodo
                        nodo.father.right.left = None
                        nodo.father.right.right = None
                        nodo.father.right.left = copia_left
                        print(f"| {nodo.data} | Eliminado")
                    
```
